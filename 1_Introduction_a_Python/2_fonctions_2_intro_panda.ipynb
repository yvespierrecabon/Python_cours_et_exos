### Importation de donn√©es avec la librairie Pandas

```python
import pandas as pd
path_csv = 'weight_height_dataset.csv'
# path_csv = './weight_height_dataset.csv'
population = pd.read_csv(path_csv)
population.head(10)
```

```python
population_height = population["Height"]
population_weight = population["Weight"]
population_height.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(population_height*2.54,population_weight,s=1)
plt.show()
```

```python
import numpy as np

fig, ax = plt.subplots()
num_bins = 100

counts, bins = np.histogram(population_height,
                            np.linspace(min(population_height),
                                       max(population_height),
                                        num_bins))
ax.hist(population_height,
       bins[:-1],
       facecolor='orange',
        weights=(1/sum(counts))*np.ones_like(population_height),
        label="Height's distribution"
       )
plt.xlabel("Taille(Inches)")

plt.show()
```

```python
print("Moyenne :",np.mean(population_height),"inch ou",
      np.mean(population_height) *2.54,"cm")
print("Variance :", np.var(population_height),"inch")
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from math import sqrt

fig, ax = plt.subplots()
num_bins = 100
population_height = population["Height"]
population_weight = population["Weight"]
population_height.head()
height_repartition = np.linspace(min(population_height),
                                       max(population_height),
                                        num_bins)
counts, bins = np.histogram(population_height, height_repartition)
ax.hist(population_height,
       bins[:-1],
       facecolor='orange',
        weights=(1/sum(counts))*np.ones_like(population_height),
        label="Height's distribution"
       )
plt.xlabel("Taille(Inches)")

dist_normale = stats.norm.rvs(loc = np.mean(population_height), scale = sqrt(np.var(population_height)), size = len(population_height))
ax.hist(dist_normale,
       bins[:-1],
       facecolor='blue',
        alpha = 0.5,
        weights=(1/sum(counts))*np.ones_like(population_height),
        label="Normal distribution"
       )

plt.show()
```
