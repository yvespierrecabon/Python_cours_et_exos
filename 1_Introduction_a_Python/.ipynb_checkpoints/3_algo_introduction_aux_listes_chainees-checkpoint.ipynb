## introduction aux listes chainées

```python
class Maillon:
    def __init__(self, v, suivant=None):
        self.valeur = v
        self.suivant = suivant
        

class Liste:
    def __init__(self):
        self.tete = None # liste vide
        
    def est_vide(self):
        return self.tete == None
    
    def ajoute_en_tete(self,m):
        if self.est_vide():
            self.tete=m
        else:
            m.suivant = self.tete
            self.tete = m
    
    def ajoute_en_tete_valeur(self,v):
        m = Maillon(v)
        if self.est_vide():
            self.tete= m
        else:
            m.suivant = self.tete
            self.tete = m
    
    def ajoute_en_queue(self, m):
        if self.est_vide():
            self.tete=m
        else:
            m_courant = self.tete
            while m_courant.suivant is not None:
                m_courant = m_courant.suivant
            m_courant.suivant = m
            
            
    def ajoute_en_queue_rec(self, m):
        def ajoute_maillon_en_queue(m_courant:Maillon, ma:Maillon):
            if m_courant is None:
                m_courant = ma
            else:
                ajoute_maillon_en_queue(m_courant.suivant, ma)
        ajoute_maillon_en_queue(self.tete,m)
            
    def ajoute_en_queue_rec_2(self, m):
        if self.est_vide():
            self.tete=m
        elif self.tete.suivant is None:
            self.tete.suivant = m
        else:
            L_suivante = Liste()
            L_suivante.ajoute_en_tete(self.tete.suivant)
            L_suivante.ajoute_en_queue_rec_2( m)
            
    def affiche(self):
        if not self.est_vide():
            m_courant = self.tete
            while m_courant is not None:
                print(m_courant.valeur)
                m_courant = m_courant.suivant
                
    def affiche_2(self):
        def affiche_maillon(m:Maillon):
            if m is not None:
                print(m.valeur)
                affiche_maillon(m.suivant)
        affiche_maillon(self.tete)
        
        
    def affiche_3(self):
        if not self.est_vide():
            print(self.tete.valeur)
            L_suivante = Liste()
            L_suivante.ajoute_en_tete(self.tete.suivant)
            L_suivante.affiche_3()        
    
    def affiche_2_inverse(self):
        def affiche_maillon(m:Maillon):
            if m is not None:
                affiche_maillon(m.suivant)
                print(m.valeur)
        affiche_maillon(self.tete)  
        
        
    """insère valeur après le maillon numéro indice.
    Le 1° maillon a pour indice 1
    si indice = 0 => insère en tête
    si indice > nb de maillon => insère en queue"""    
    def insere_apres_indice(self, indice, valeur):
        if indice == 0:
            self.ajoute_en_tete_valeur(valeur)
        else:
            i = 1
            maillon_nouveau = Maillon(valeur)
            maillon_courant = self.tete
            while i<indice and maillon_courant.suivant is not None:
                maillon_courant = maillon_courant.suivant
                i +=1
            maillon_nouveau.suivant = maillon_courant.suivant
            maillon_courant.suivant = maillon_nouveau
            
    """
    1° maillon a pour indice 1
    si indice > nb de maillon, return false
    si ok, return True
    """    
    def supprime_indice(self, indice):
        if indice <1 or self.est_vide():
            return False
        elif indice == 1:
            self.tete =  self.tete.suivant
            return True
        else:    
            i = 2
            maillon_courant = self.tete
            while i<indice and maillon_courant.suivant is not None:
                maillon_courant = maillon_courant.suivant
                i +=1
            if i == indice:
                maillon_courant.suivant = maillon_courant.suivant.suivant
                return True
            else:
                return False
            
            
    def supprime_valeur(self, valeur):
        if self.est_vide():
            return False
        
            
        pass
```

### Classe Maillon sans classe Liste, mais avec fcts récursives

```python
class Maillon():

    def __init__(self, val, suivant = None):
        self.val = val
        self.suivant = suivant

def insert_en_tete(lst:Maillon, val):
    if lst is None:
        return Maillon(val)
    else:
        return Maillon(val, lst)

def est_vide(lst:Maillon):
    return lst is None

def affiche_chaine(lst:Maillon):
    if not est_vide(lst):
        temp = lst
        while temp != None:
            print(temp.val)
            temp = temp.suivant

def affiche_chaine_2(lst:Maillon):
    if not est_vide(lst):
        print(lst.val)
        affiche_chaine_2(lst.suivant)

def affiche_chaine_reverse(lst:Maillon):
    if not est_vide(lst):
        affiche_chaine_reverse(lst.suivant)
        print(lst.val)

def insert_en_queue(lst:Maillon, val):
    if est_vide(lst):
        return Maillon(val)
    else:
        lst.suivant = insert_en_queue(lst.suivant, val)
        return lst

def longueur_chaine(lst:Maillon):
    if est_vide(lst):
        return 0
    else:
        return 1 + longueur_chaine(lst.suivant)

"""
premier maillon indice 1
si indice superieur a longueur, alors inserer en queue 
"""
def insert_apres(lst:Maillon, indice, val):
    if indice == 1 or lst.suivant is None:
        m = Maillon(val, lst.suivant)
        lst.suivant = m
    else:
        insert_apres(lst.suivant, indice - 1, val)

def supprime(lst:Maillon, val):
    if lst is not None:
        if lst.val == val:
            print("val:", val)
            return lst.suivant
        else:
            lst.suivant = supprime(lst.suivant, val)
            return lst

#inverser,
maillon = Maillon(0)
for i in range(1, 10):
    maillon = insert_en_tete(maillon, i)

#affiche_chaine(maillon)
affiche_chaine_2(maillon)

maillon = insert_en_queue(maillon, 10)
print("")
affiche_chaine_2(maillon)

print(longueur_chaine(maillon))

#print("")
#affiche_chaine_reverse(maillon)

print("")
insert_apres(maillon, 5 , 100)
affiche_chaine_2(maillon)

print("")
insert_apres(maillon, 100 , 200)
affiche_chaine_2(maillon)

print("")
insert_apres(maillon, 0 , 50)
affiche_chaine_2(maillon)

print("")
maillon = supprime(maillon, 100)
affiche_chaine_2(maillon)
```
