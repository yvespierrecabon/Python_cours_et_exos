```python
def table_de_multiplication(n):
    print('Table de multiplication de',n)
    for i in range(n+1):
        print(i,'*',n,'=',i*n)

table_de_multiplication(7)
```

### variables locales, globales

```python
def ajouter_1(x):
    x = x+1
    return x

x = 10
print('x =',x)

print('valeur retournée par ajouter_1(x) :',ajouter_1(x))
print('x =',x)

a = 0
print('a =',a)
print('valeur retournée par ajouter_1(a) :',ajouter_1(a))
print('a =',a)

def ajouter_1_a_x():
    global x
    x= x+1
    return x
print('x =',x)
print('valeur retournée par ajouter_1_a_x(a) :',ajouter_1_a_x())
print('x =',x)
```

```python
def somme_produit(a,b):
    return a+b, a*b


s,p = somme_produit(2,6)
print(s,p)
```

```python
def n_premiers_carres(n):
    return [a**2 for a in range(n+1)]


print(n_premiers_carres(10))
```

### Fonctions recursives

```python
def factoriel_iter(n):
    res = 1
    for i in range(2,n+1):
        res*=i
    return res

print(factoriel_iter(5))
```

```python
def factoriel_recursif(n):
    if n == 1:
        return 1
    else:
        return n * factoriel_recursif(n-1)


print(factoriel_recursif(5))
```
