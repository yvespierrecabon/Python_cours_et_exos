### Fct récursives retournant des valeurs

```python
def somme_des_n_premiers_entiers(n):
    if n <= 0:
        return 0
    else:
        return n + somme_des_n_premiers_entiers(n-1)
    
for n in range(10):
    print('somme_des_n_premiers_entiers(',n,') = ',somme_des_n_premiers_entiers(n),sep='')
```

```python
def produit_des_n_premiers_entiers(n):
    if n <= 0:
        return 1
    else:
        return n * produit_des_n_premiers_entiers(n-1)
    
for n in range(10):
    print('produit_des_n_premiers_entiers(',n,') = ',produit_des_n_premiers_entiers(n),sep='')
    
```

```python
def Fibonacci(n):
    if n == 0 or n == 1:
        return n
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)
    
for n in range(2, 10):
    print('Fibonacci(',n,') = ',Fibonacci(n),sep='')
   
```

### Exemple de récursives

- Afficher toutes les années entre 2 années

```python
def affiche_annees_rec(debut,fin):
    if debut <= fin:
        print(debut)
        affiche_annees_rec(debut+1,fin)
    
affiche_annees_rec(2000,2010)
```

- fct puissance

```python
def puissance_rec(x, n):
    if n == 1:
        return x
    else:
        return x * puissance_rec(x, n-1)
    
print('2 puissance 3 =',puissance_rec(2,3))
print('3 puiss 2 =',puissance_rec(3,2))
```

### fonction récursive

```python
from matplotlib import pyplot as plt
import math
def factorielle_iterative_1(n):
    res = 1
    for i in range(2,n+1):
        res = res*i
    return res


def factorielle_iterative_2(n):
    res = 1
    i = 2
    while(i<=n):
        res = res*i
        i+=1
    return res


def factorielle_recursive_1(n):
    if n==1:
        return 1
    else:
        return n*factorielle_recursive_1(n-1)
    
    
def factorielle_recursive_2(n):
    if n!=1:
        return n*factorielle_recursive_2(n-1)
    else:
        return 1

test = 5
print('factorielle_iterative_1(',test,') =',factorielle_iterative_1(test))
print('factorielle_iterative_2(',test,') =',factorielle_iterative_2(test))      
print('factorielle_recursive_1(',test,') =',factorielle_recursive_1(test))
print('factorielle_recursive_2(',test,') =',factorielle_recursive_2(test))


min = 2
max =50
nb_tests= 100
from time import time


X = [x for x in range(nb_tests)]
fr_1=[]
for x in range(nb_tests):
    start = time()
    for i in range (min+x, max+x):
        f = factorielle_recursive_1(i)
        # print('factorielle_recursive',i,'=',f)
    stop = time()
    fr_1.append(stop - start)
    # print('temps recursive : ',stop - start)

plt.plot(X,fr_1,label='recursive 1')
fr_1.clear()


for x in range(nb_tests):
    start = time()
    for i in range (min+x, max+x):
        f = factorielle_recursive_2(i)
        # print('factorielle_recursive_2',i,'=',f)
    stop = time()
    # print('temps recursive 2: ',stop - start)
    fr_1.append(stop - start)
plt.plot(X,fr_1,label='recursive 2')
fr_1.clear()

for x in range(nb_tests):
    start = time()
    for i in range (min+x, max+x):
        f = factorielle_iterative_1(i)
        # print('factorielle',i,'=',f)
    stop = time()
    # print('temps itérative: ',stop - start)
    fr_1.append(stop - start)
plt.plot(X,fr_1,label='itérative 1')
fr_1.clear()

for x in range(nb_tests):
    start = time()
    for i in range (min+x, max+x):
        f = factorielle_iterative_2(i)
        # print('factorielle',i,'=',f)
    stop = time()
    # print('temps itérative: ',stop - start)
    fr_1.append(stop - start)
plt.plot(X,fr_1,label='itérative 2')
fr_1.clear()


for x in range(nb_tests):
    start = time()
    for i in range (min+x, max+x):
        f = math.factorial(i)
        # print('factorielle',i,'=',f)
    stop = time()
    # print('temps itérative: ',stop - start)
    fr_1.append(stop - start)
plt.plot(X,fr_1,label='factorial')
fr_1.clear()

plt.legend()
plt.show()
```

### Triangles en fractale


```python
from turtle import *
def fractale(deep, length,poly, colors):
    if deep >= 1:
        begin_fill()
        for i in range(poly):
            color = '#'+str(hex(randint(1,255)))[2:].zfill(2)+str(hex(randint(1,255)))[2:].zfill(2)+str(hex(randint(1,255)))[2:].zfill(2)
            # fillcolor(colors[deep%len(colors)])
            fillcolor(color)
            forward(length)
            left(360/poly)

            fractale(deep-1, length/2,poly,colors)
        end_fill()
        
tableau_couleur = ('red','blue','green','black', "orange")
speed("fastest")
width(1)
poly=3
up()
sety(-200)
down()
fractale(5,400,poly,tableau_couleur)
exitonclick()
```

#### flocon de Koch

```python
from turtle import *
def fractale_2(deep, length):
    if deep >=1:
        fractale_2(deep-1, length/3)
        left(45)
        fractale_2(deep-1, length/3)
        right(90)
        fractale_2(deep-1, length/3)
        left(45)
        fractale_2(deep-1, length/3)
    else:
        forward(length)
        
speed(0)
fractale_2(3,500)
exitonclick()
```

#### Courbe de Hilbert

```python
from turtle import *

def fractale_3(angle, deep):
    if deep >=1:
        right(angle)
        fractale_3(-angle,deep-1)
        forward(l)
        left(angle)
        fractale_3(angle, deep-1)
        forward(l)
        fractale_3(angle, deep-1)
        left(angle)
        forward(l)
        fractale_3(-angle, deep-1)
        right(angle)

        
speed(0)
l = 10
fractale_3(90,5)
exitonclick()
```
